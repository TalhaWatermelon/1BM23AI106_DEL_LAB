```python
import numpy as np
import matplotlib.pyplot as plt

# Linear function
def linear(x):
    return x


x = np.linspace(-10, 10, 400)
y = linear(x)
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Linear", color='red')
plt.title("Linear Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()




import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 400)
y = sigmoid(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Sigmoid", color='blue')
plt.title("Sigmoid Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()



import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

x = np.linspace(-10, 10, 400)
y = tanh(x)


plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Tanh", color='red')
plt.title("Tanh Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()


import numpy as np
import matplotlib.pyplot as plt

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Generate input values
x = np.linspace(-10, 10, 400)
y = relu(x)

# Plot ReLU function
plt.figure(figsize=(8, 5))
plt.plot(x, y, label="ReLU", color='green')
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()



import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x / exp_x.sum(axis=0)

x1, x2 = 1.0, 2.0
vary = np.linspace(-5, 5, 400)
softmax_outputs = []

for v in vary:
    inp = np.array([x1, x2, v])
    sm = softmax(inp)
    softmax_outputs.append(sm)

softmax_outputs = np.array(softmax_outputs)

plt.figure(figsize=(8,5))
plt.plot(vary, softmax_outputs[:,0], label='Softmax output for x1=1.0')
plt.plot(vary, softmax_outputs[:,1], label='Softmax output for x2=2.0')
plt.plot(vary, softmax_outputs[:,2], label='Softmax output for varying input')
plt.title('Softmax Output Probabilities for 3-class Input')
plt.xlabel('Varying third input value')
plt.ylabel('Softmax Probability')
plt.legend()
plt.grid(True)
plt.show()



import numpy as np
import matplotlib.pyplot as plt

def linear(x):
    return x

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, alpha * x)

def softmax(x):
    e_x = np.exp(x - np.max(x, axis=0))
    return e_x / e_x.sum(axis=0)

x = np.linspace(-10, 10, 400)

plt.figure(figsize=(10, 6))
plt.plot(x, linear(x), label="Linear", color='red')
plt.plot(x, sigmoid(x), label="Sigmoid", color='blue')
plt.plot(x, tanh(x), label="Tanh", color='orange')
plt.plot(x, relu(x), label="ReLU", color='green')
plt.plot(x, leaky_relu(x), label="Leaky ReLU", color='purple')

vec = np.array([x, 2*x, -x])
softmax_vals = softmax(vec)

plt.plot(x, softmax_vals[0], label="Softmax class 1", linestyle='--', color='brown')
plt.plot(x, softmax_vals[1], label="Softmax class 2", linestyle='--', color='pink')
plt.plot(x, softmax_vals[2], label="Softmax class 3", linestyle='--', color='gray')

plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend(loc='upper left')
plt.show()


```
